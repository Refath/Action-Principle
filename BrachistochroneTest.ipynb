```python
from vpython import *
```

```python
scene = canvas(background = color.white)
scene.range = 2
boundaryCond = []
dots = []

c = curve(radius = 0.05)

def showSphere(evt):
    loc = evt.pos
    boundaryCond.append(sphere(pos=loc, radius=0.2, color=color.cyan))

scene.bind('click', showSphere)

while len(boundaryCond)<2:
    rate(5)

scene.unbind('click', showSphere)

c.append(boundaryCond[0].pos)

def makeCurve(evt):
    c.append(evt.pos)
    
scene.bind('click', makeCurve)

def getevent():
    obj = scene.mouse.pick
    
    if (obj == boundaryCond[0]):
        obj.color = color.green
        
    if (obj == boundaryCond[1]):
        obj.color = color.red
        c.append(obj.pos)
        dots.append(obj.pos)
        data = c.slice(0,len(dots))
     
        ball = sphere(pos = boundaryCond[0].pos+0.1*norm(data[1].pos-data[0].pos), radius = 0.1, color = color.yellow)
        
        t = 0
        dt = 0.01
        
        tgraph=graph(title="Particle Motion", xtitle="Time [s]", fast=False, align = 'right', width=600, height=400)
        f2 = gdots(color=color.green, label = 'Y Position')
        f3 = gdots(color=color.red, label = 'Velocity')
        
        fgraph=graph(title="Path of Least Action", xtitle="Time [s]", fast=False, align = 'right', width=600, height=400)
        f4 = gdots(color=color.blue, label = 'Kinetic Energy KE')
        f5 = gdots(color=color.cyan, label = 'Potential Energy PE')
        f6 = gdots(color=color.magenta, label = 'Action (KE-PE)')
     
        for i in range(1,len(data)):
            while ball.pos.x < data[i].pos.x: 
                rate(50)
                ball.v = vec(0,0,0)
                ball.v += norm(data[i].pos - data[i-1].pos) * sqrt(2 * 9.8 * boundaryCond[0].pos.y-ball.pos.y)
                ball.pos += ball.v * dt
               
                f2.plot(t,ball.pos.y)
                f3.plot(t,mag(ball.v))
                
                KineticEnergy = 0
                PotentialEnergy = 0
                
                KineticEnergy += 0.5 * 5 * (mag(ball.v) ** 2)
                PotentialEnergy += 5 * 9.8 * (boundaryCond[0].pos.y-ball.pos.y)
                Action = KineticEnergy - PotentialEnergy
                
                f4.plot(t,KineticEnergy)
                f5.plot(t,PotentialEnergy)
                f6.plot(t,Action)
                
                t += dt
            
scene.bind("mousedown", getevent)

```
